```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import pandas as pd
import re

# Function to extract webpage link from Google Maps place page
def extract_webpage_link(driver, place_url):
    try:
        # Open the Google Maps place page
        driver.get(place_url)
        time.sleep(3)  # Add a short delay to ensure the page fully loads (adjust as needed)

        # Find all anchor elements on the page
        anchor_elements = driver.find_elements(By.TAG_NAME, "a")

        # Initialize variable to store webpage link
        webpage_link = None

        # Regex pattern to match a valid URL
        url_pattern = r'(http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+)'

        # Loop through anchor elements to find any valid webpage link
        for anchor in anchor_elements:
            href = anchor.get_attribute("href")
            if href and "https://food.google.com/" not in href:
                match = re.search(url_pattern, href)
                if match:
                    webpage_link = match.group().strip()
                    break  # Stop further processing once webpage link is found

        return webpage_link

    except Exception as e:
        print(f"Error processing {place_url}: {str(e)}")
        return None

# Configure Selenium Chrome options
chrome_options = Options()
chrome_options.add_argument("--lang=ja")  # Set language to Japanese if needed
chrome_options.add_argument("--headless")  # Optional: Run Chrome in headless mode
service = Service('/home/tomamaya/Downloads/chromedriver')  # Replace with actual path to chromedriver

# Initialize WebDriver
driver = webdriver.Chrome(service=service, options=chrome_options)

# Example list of URLs (replace with your own list or read from DataFrame)
urls = [
    "https://www.google.com/maps/place/Hacienda+del+cielo+MODERN+MEXICANO/data=!4m7!3m6!1s0x60188b454758150f:0x911db2c2dfc8b756!8m2!3d35.6504276!4d139.7025852!16s%2Fg%2F1wyc49wb!19sChIJDxVYR0WLGGARVrfI38KyHZE?authuser=0&hl=ja&rclk=1",
    "https://www.google.com/maps/place/%E3%82%BF%E3%82%B3%E3%83%99%E3%83%AB+%E6%B8%8B%E8%B0%B7%E9%81%93%E7%8E%84%E5%9D%82%E5%BA%97/data=!4m7!3m6!1s0x60188ca9be0b957f:0xa34dd8d7530b5518!8m2!3d35.6599248!4d139.6976752!16s%2Fg%2F11b7jftrn7!19sChIJf5ULvqmMGGARGFULU9fYTaM?authuser=0&hl=ja&rclk=1"
]

# Iterate through each URL and extract webpage link
for url in urls:
    webpage_link = extract_webpage_link(driver, url)
    if webpage_link:
        print(f"Webpage link for {url}: {webpage_link}")
    else:
        print(f"Webpage link not found for {url}")

# Close the WebDriver
driver.quit()
```
